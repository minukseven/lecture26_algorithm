### 소수(Prime Number)

```python
def is_prime_number(N):
    if N == 1:
        return False

    is_prime = True
    for i in range(2, N):
        if N % i == 0:
            is_prime = False
            break
    
    return is_prime
```

```python
print(is_prime_number(5))
```

```python
def is_prime_number(N):
    if N == 2:
        return True
    if N == 1 or N % 2 == 0:
        return False

    for i in range(3, int(N**0.5) + 1, 2):
        if N % i == 0:
            return False

    return True

print(is_prime_number(7))   
print(is_prime_number(10))
```

```python
def is_prime_v1(N):
    if N == 2:
        return True
    if N == 1 or N % 2 == 0:
        return False

    for i in range(3, N, 2):
        if N % i == 0:
            return False

    return True


# 실행
N = int(input("숫자 입력: "))
print("1단계:", f"{N}은 소수입니다." if is_prime_v1(N) else f"{N}은 소수가 아닙니다.")
```

```python
def is_prime_v2(N):
    if N == 2:
        return True
    if N == 1 or N % 2 == 0:
        return False

    for i in range(3, N // 2 + 1, 2):
        if N % i == 0:
            return False

    return True


# 실행
N = int(input("숫자 입력: "))
print("2단계:", f"{N}은 소수입니다." if is_prime_v2(N) else f"{N}은 소수가 아닙니다.")
```

```python
def is_prime_v3(N):
    if N == 2:
        return True
    if N == 1 or N % 2 == 0:
        return False

    primes = []

    for i in range(2, N):
        is_prime = True

        for p in primes:
            if i % p == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(i)

        if N % i == 0:
            return False

    return True


# 실행
N = int(input("숫자 입력: "))
print("3단계:", f"{N}은 소수입니다." if is_prime_v3(N) else f"{N}은 소수가 아닙니다.")
```
