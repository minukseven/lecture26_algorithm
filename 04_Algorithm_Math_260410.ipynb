### 소수(Prime Number)

```python
def is_prime_number(N):
    if N == 1:
        return False

    is_prime = True
    for i in range(2, N):
        if N % i == 0:
            is_prime = False
            break
    
    return is_prime
```

```python
print(is_prime_number(5))
```

```python
def is_prime_number(N):
    if N == 2:
        return True
    if N == 1 or N % 2 == 0:
        return False

    for i in range(3, int(N**0.5) + 1, 2):
        if N % i == 0:
            return False

    return True

print(is_prime_number(7))   
print(is_prime_number(10))
```
